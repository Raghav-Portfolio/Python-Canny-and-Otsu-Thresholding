```python
import cv2
import matplotlib.pyplot as plt
```

```python
# reading the image
image = cv2.imread(r"D:\task_1\coin1.jpeg")
```

```python
# converting to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
#converting to grayscale
gray_img = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
# creating a binary thresholded image 
_, binary = cv2.threshold(gray_img, 0, 255,cv2.THRESH_BINARY+cv2.THRESH_OTSU)
```

```python
# finding the contours from the thresholded image
contours, hierarchy = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_NONE)
#drawing all contours
image1 = cv2.drawContours(image, contours, -1, (0, 255, 0), 2)
print('Number of Contours:'+ str(len(contours)))
plt.imshow(image1)
plt.show()
```

```python
x=int(input('please enter the x coordinate'))
y=int(input('please enter the y coordinate'))
point1=(x,y)
a=cv2.pointPolygonTest(contours[-1],(point1), False)
cv2.circle(image1, point1, 4, (100, 100, 255), -1)
cv2.putText(image1, 'point', (point1[0] -10, point1[1] -20), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 0, 0), lineType=cv2.LINE_AA)
plt.imshow(image1)
plt.show()
if a<0:
   print('outside')
elif a>0: 
   print('inside')
else:
   print('on the contour')
```
